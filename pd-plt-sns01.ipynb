```python
#Cruzando dados
Parte01 = bd[['NU_INSCRI','NO_MUNICIPIO_PROVA']].head(6)
Parte02 = bd[['NU_INSCRI','NU_IDADE','TP_SEXO']].head()
print(Parte01)
print('\n')
print(Parte02)
print('\n')

#Inner
pd.merge(Parte01,Parte02,how='inner',on=['NU_INSCRI'])
#Outer
pd.merge(Parte01,Parte02,how='outer',on=['NU_INSCRI'])
#Right
pd.merge(Parte01,Parte02,how='right',on=['NU_INSCRI'])
#Left
pd.merge(Parte01,Parte02,how='left',on=['NU_INSCRI'])
```

```python
#Cruzando dados
Parte01 = bd[['NU_INSCRI','NO_MUNICIPIO_PROVA']].iloc[0:5,0:1]
Parte02 = bd[['NU_INSCRI','NO_MUNICIPIO_PROVA']].iloc[5:10,0:1]
Parte03 = bd[['NU_INSCRI','NO_MUNICIPIO_PROVA']].iloc[15:20,0:1]
print(Parte01)
print('\n')
print(Parte02)
print('\n')
print(Parte03)
print('\n')

pd.concat([Parte01,Parte02,Parte03]).reset_index(drop=True)
```

```python
# Criando funções
def Editar_cor_raca(cor_raca):

    if cor_raca == 0:
        return 'Não informado'

    elif cor_raca == 1:
        return 'Branca'

    elif cor_raca == 2:
        return 'Preta'

    elif cor_raca == 3:
        return 'Parda'

    elif cor_raca == 4:
        return 'Amarela'

    else:
        return 'Indígena'

bd['Aplicando a função'] = bd['TP_COR_RACA'].apply( Editar_cor_raca )

bd.head()
```

```python
# Gerar gráficos com o pandas

# Boxplot
Dados = bd.dropna()[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT']]
Dados.plot.box();
```

```python
# Gráfico de Barras
Analise = bd.groupby(['Aplicando a função']).mean()['NU_IDADE'].reset_index()

Analise.plot.bar( x='Aplicando a função', y='NU_IDADE', alpha=0.5 );
```
